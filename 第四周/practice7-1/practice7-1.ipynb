```python
import matplotlib.pyplot as plt
from PIL import Image
plt.rcParams["font.family"]="SimHei"

img = Image.open("lena.tiff")
img_r,img_g,img_b=img.split()
plt.figure()

plt.subplot(221)
plt.title("R-",fontsize="14")
img_small=img_r.resize((50,50))
plt.imshow(img_small)

plt.subplot(222)
plt.title("G-+",fontsize="14")
img_x=img_g.transpose(Image.FLIP_LEFT_RIGHT)
img_trans=img_x.transpose(Image.ROTATE_90)
plt.imshow(img_trans)

plt.subplot(223)
plt.axis("off")
plt.title("B-",fontsize="14")
img_region=img_b.crop((0,0,150,150))
plt.imshow(img_region)

img_rgb=Image.merge("RGB",[img_r,img_g,img_b])
plt.subplot(224)
plt.axis("off")
plt.title(img.mode)
plt.imshow(img_rgb)

img_rgb.save("test.png")
plt.suptitle("", fontsize="20", color="blue")

plt.tight_layout(rect=[0,0,1,0.9])

plt.show()
```
